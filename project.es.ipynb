#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
# TODO
import matplotlib.pyplot as plt

subset_ds = ds[ds['level5'].isin(poblaciones)]
df = pd.DataFrame(subset_ds, columns=['level5', 'price']) 

min_price = df['price'].min()
dif_price = df['price'].max() - min_price
df.loc[:, 'normalize_price'] = (df['price'] - min_price) / dif_price

Sector = df.groupby('level5')

plt.figure(figsize=(10, 6))
for key, item in Sector:
    plt.hist(item['normalize_price'], bins=20, alpha=0.7, label=key)



plt.xlabel('Precio Normalizado')
plt.ylabel('Frecuencia')
plt.title('Distribución de Precios Normalizados para Ciudades Seleccionadas')
plt.legend()
plt.show()

```

Ejecuta la siguiente celda y luego comienza a codear en la siguiente. Debes implementar u código simple que transforme las columnas de coordenadas en un diccionario de python (agrega más información si es necesario) y agrégala al mapa.

```python
from ipyleaflet import Map, basemaps, Marker, Icon # type: ignore

# Mapa centrado en (60 grados latitud et -2.2 grados longitud)
# Latitud, longitud
map = Map(center = (60, -2.2), zoom = 2, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.OpenStreetMap.Mapnik)
map
```

```python
## Aquí: traza la coordenadas de los estados

## PON TU CÓDIGO AQUÍ:
subset_ds = ds[ds['level5'].isin(poblaciones)]
df = pd.DataFrame(subset_ds, columns=['level5', 'latitude', 'longitude']) 

icons = {
    "Fuenlabrada": "https://e7.pngegg.com/pngimages/588/276/png-clipart-red-button-computer-icons-drawing-red-dot-s-orange-area.png",
    "Leganés": "https://e7.pngegg.com/pngimages/178/851/png-clipart-blue-dot-logo-resurs-bank-blue-ball-logo-icons-logos-emojis-bank-logos-thumbnail.png",
    "Getafe": "https://w7.pngwing.com/pngs/980/921/png-transparent-green-dot-corporation-business-cercle-de-fermieres-d-ahuntsic.png",
    "Alcorcón": "https://w7.pngwing.com/pngs/639/582/png-transparent-art-service-espn-gold-dots-miscellaneous-service-orange.png"
    }

for index, row in df.iterrows():
    icon_url = icons[row['level5']]
    icon = Icon(icon_url=icon_url, icon_size=[24, 24])
    marker = Marker(location=(row['latitude'], row['longitude']), icon=icon)
    map.add_layer(marker)

print(map)

```
